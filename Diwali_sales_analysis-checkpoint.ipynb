```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Open CSV file 
df = pd.read_csv(r"C:\Users\kaush\Downloads\Diwali Sales Data.csv", encoding = "ISO-8859-1")
```

```python
df.shape
```

```python
df.info()
```

```python
## Drop columns with NaN values (completely Blank)

df.drop(columns = ["Status","unnamed1"], inplace = True)
```

```python
df.isna().sum()
```

```python
# Checking NAN values for column --Amount--

df[df["Amount"].isna()]
```

```python
# drop rows having NaN values

df.dropna(inplace = True)
```

```python
# verifying NaN values again

df.isna().sum()

# No NaN value found
```

```python
# Will create new column with slabs of Age groups 


df["Age Group"] = pd.cut(df.Age, bins = [12,25,40,55,70,85,92], labels = ["10-25","25-40","40-55","55-70","70-85","85-92"])
df.head()
```

```python
df.describe()
```

# All the neccessary data cleaning is completed, so now we will proceed for EDA

```python
# Male/Female Ratio of total orders placed
gender_count = df["Gender"].value_counts()
gender_count
```

```python
# Sum of orders by Age Group
Age_grou = df["Age Group"].value_counts()
```

```python
# Total 10 States based on Sales 
top_10_states = df.groupby('State').Amount.sum().sort_values(ascending = False).head(10)/1000000
Sale_mil = top_10_states.round(2)
Sale_mil
```

```python
# Zone wise total sales
a = df.groupby("Zone").Amount.sum().sort_values(ascending = False)/1000000
Zone_sal = a.round(2)
Zone_sal
```

### Creating a Dashboard with neccessary visuals for Analysing the data

```python
plt.figure(figsize = (16,10), dpi = 200 , facecolor = "cyan", edgecolor = "red" ,  linewidth = 10 , frameon = True ,
layout = "tight")
plt.suptitle("DASHBOARD", color = "red", fontsize = 30 , fontweight = 100)
plt.subplot(2,2,1)
gender_count.plot.pie(autopct = '%1.1f%%', startangle = 90, labels = ["Female","Male"])
plt.title("Gender_distribution in %", fontsize = 20)
plt.ylabel("")
plt.subplot(2,2,2)
ax = Age_grou.plot(kind = "bar" , color = "b", width = 0.8)
plt.title('AGE_GROUP WISE COUNT', fontsize = 20)
plt.xlabel("AGE_GROUP",fontsize = 15)
plt.ylabel("COUNT", fontsize = 15)
plt.xticks(rotation = 45)
for i in ax.containers:
    ax.bar_label(i, label_type = "edge")
ax.set_facecolor("cyan")
plt.subplot(2,2,3)
top = Sale_mil.plot(kind = "bar", color = 'b', width = 0.8)
top.set_facecolor("cyan")
plt.title("TOP 10 STATE BY SALE (Million)", fontsize = 20)
plt.xlabel("STATE",fontsize = 15)
plt.ylabel("SALE",fontsize = 15)
plt.xticks(rotation = 45)
for i in top.containers:
    top.bar_label(i,label_type = "edge")
plt.subplot(2,2,4)
Zone_sal.plot(kind = "pie",startangle = 90, autopct = "%1.1f%%")
plt.title("ZONE WISE SALE", fontsize = 20)
plt.ylabel("")
    
plt.show()

```

#### 1) From the above Dashboard we can conclude that Female ratio of orders is higher than male -- approx - 70 %
#### 2) Age group between 25-40 is placing maximum orders.
#### 3) Most revenue is generated from UP state.
#### 4) Central Zone is having the maximum revenue.

```python
## Sales across different working sectorsl
plt.figure(figsize = (20,4))
Occu_sal = sns.countplot(data = df , x = "Occupation",palette= "viridis")
for i in Occu_sal.containers:
    Occu_sal.bar_label(i)
```

#### From above bar graph we conclude that most orders are placed from IT,Healthcare and Aviation sector

```python
## Sale done for different types of products_Categories -- TOP 10 HIGH SELLING
plt.figure(figsize = (20,5))
a = df.groupby("Product_Category").Amount.sum().sort_values(ascending = False)/100000
Pro_sal = a.round(2).reset_index().head(10)
Pro_sal
ax = sns.barplot(data = Pro_sal, x = "Product_Category", y = "Amount")
plt.title("TOTAL SALE BY PRODUCT_CATEGORY (In Lacs)", fontsize = 20)
plt.xlabel("PRODUCTS",fontsize = 15)
plt.ylabel("TOTAL SALE",fontsize = 15)
for i in  ax.containers:
    ax.bar_label(i)

```

#### From above bar graph we can conclude that people are spending the most on Food, clothes and electronics
